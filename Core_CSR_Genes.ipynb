# Find the "core stress response genes" meaning the genes that are common between all five of the different types of cellular stress responses

---



```python
# Create a list of the five dataframes
responses_list = [DDR, PDR, oxidative, heat, hypoxia]

# Concatenate the DBID columns from all DataFrames into one Series
concatenated_ids = pd.concat([df['DBID'] for df in responses_list], ignore_index=True)

# Count the occurrences of each DBID
id_counts = concatenated_ids.value_counts()

# Filter DBIDs that appear in all dataframes
core_ids_all = id_counts[id_counts >= 5].index

# Initialize an empty DataFrame to store the common DBIDs and Names
core_genes = pd.DataFrame(columns=['DBID', 'NAME'])

# Iterate through each DataFrame, filter for common DBIDs, and concatenate
for df in responses_list:
    core_genes = pd.concat([core_genes, df[df['DBID'].isin(core_ids_all)]], ignore_index=True)

# Sort the DataFrame by 'DBID'
core_genes.sort_values(by='DBID', inplace=True)

core_genes.head(15)
```

We see that the previous dataframe has 350 rows, and every unique ID is repeated five times. That is because the other columns  ("TOTAL_FREQUENCY" and "NUM_ARTICLES") have different values in each of the five dataframes.
Below we will keep in a dataframe only the unique IDs without any duplicates.


```python
# Drop duplicate rows based on DBID only
unique_core_genes = core_genes.drop_duplicates(subset=['DBID'])
unique_core_genes
```

# Find the unique genes extracted from Darling for each of the five types of cellular stress response

---



```python
# Create a directory of the five dataframes
dataframes_dir = {'oxidative': oxidative, 'heat': heat, 'DDR': DDR, 'PDR': PDR, 'hypoxia': hypoxia}

unique_values_per_dataframe = {}

# Iterate over each dataframe
for df_name, df in dataframes_dir.items():
    # Find unique DBIDs
    unique_dbids = df['DBID'].unique()
    unique_values = []
    for dbid in unique_dbids:
        # Find the corresponding name for each unique DBID
        name = df.loc[df['DBID'] == dbid, 'NAME'].iloc[0]
        # Check if the unique DBID is not present in other dataframes_dir
        if all(dbid not in other_df['DBID'].unique() for other_name, other_df in dataframes_dir.items() if other_name != df_name):
            unique_values.append((dbid, name))
    unique_values_per_dataframe[df_name] = unique_values

# Convert the dictionary to a dataframe
result_df = pd.DataFrame([(df_name, dbid, name) for df_name, values in unique_values_per_dataframe.items() for dbid, name in values], columns=['DataFrame', 'DBID', 'Name'])

# Display the result dataframe
print(result_df)
```

```python
# Split the dataframes per type of stress response and save them as csv files separately

# Initialize a dictionary to store the new dataframes
split_dataframes = {}

# Iterate over each unique DataFrame name in the result_df
for df_name in result_df['DataFrame'].unique():
    # Filter the result_df for the current DataFrame name
    filtered_df = result_df[result_df['DataFrame'] == df_name]
    # Drop the 'DataFrame' column as it's no longer needed
    filtered_df = filtered_df.drop(columns=['DataFrame'])
    # Store the filtered dataframe in the dictionary
    split_dataframes[df_name] = filtered_df

# Dictionary with the dataframes for each original dataframe name
for df_name, df in split_dataframes.items():
    print(f"DataFrame: {df_name}")
    print(df)
    print()
```

```python
#Export the dataframes to a google drive drirectory

output_dir = '/content/drive/MyDrive/Διπλωματική/Mesh Search/' #The google directory where the files will be saved

result_df.to_csv('/content/drive/MyDrive/Διπλωματική/Mesh Search/unique_genes_per_response.csv')

#Export the split dataframes to different csv files
for df_name, df in split_dataframes.items():
    output_path = f"{output_dir}{df_name}_unique_genes.csv"
    df.to_csv(output_path, index=False)
    print(f"CSV file for {df_name} saved to {output_path}")
```

# Find the how often and in how many articles the 70 core genes were found by Darling for each type of cellular stress response

```python
# Read the CV file containing the core stress genes
core_genes = pd.read_csv('/content/drive/MyDrive/Διπλωματική/Mesh Search/unique_core_genes_all_responses.csv', sep = ';')
```

```python
#Initialize the output dataframe
output_df = core_genes[['NAME', 'DBID']]
```

```python
#Define a function to merge the data from each dataset of the types of CSR
def merge_gene_data(df, name, output_df):
    temp_df = df[['DBID', 'TOTAL_FREQUENCY', 'NUM_ARTICLES']]
    merged_df = pd.merge(output_df, temp_df, how='left', on='DBID')
    merged_df.rename(columns={'TOTAL_FREQUENCY': f'Frequency_{name}', 'NUM_ARTICLES': f'Articles_{name}'}, inplace=True)
    return merged_df
```

```python
# Iterate through the given dataframes and merge the data
dataframes = {'DDR': DDR, 'PDR': PDR, 'OSR': oxidative, 'HSR': heat, 'HR': hypoxia}
for name, df in dataframes.items():
    output_df = merge_gene_data(df, name, output_df)
```
